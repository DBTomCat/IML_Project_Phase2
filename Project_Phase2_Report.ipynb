# Theory Question 1
If we have normal prior distribution $p_Z(z) = \mathcal{N}(µ_z, Σ_z)$ and normal likelihood distribution $ p_{Y|Z}(y|z) = \mathcal{N}(Wz + b, Σ_{Y|Z}) $, show that joint distribution will be Normal and find its parameters ($µ_{Y,Z}, Σ_{Y,Z}$).

Answer:


First we prove a lemma that we’ll use later:

Consider the below matrix M in block form, the goal is to calculate its determinant using its components (A, B, C and D).

$$
M = \begin{bmatrix}
   A & B \\
   C & D
\end{bmatrix}
$$

Imagine the matrix M has an LU decomposition such that $ M = LU $

$$
L = \begin{bmatrix}
   I & 0 \\
   L_{21} & L_{22}
\end{bmatrix} \quad , \quad U = \begin{bmatrix}
   U_{11} & U_{12} \\
   0 & U_{22}
\end{bmatrix}
$$

Where we can easily show that:

$$
U_{11} = A \quad , \quad L_{21} = CA^{-1} \quad , \quad U_{12} = B \quad , \quad L_{22}U_{22} = D - CA^{-1}B
$$

Now we can write:

\begin{gather*}
\det\left(M\right) = \det\left(LU\right) = \det\left(L\right).\det\left(U\right) \\
= \det\left(I \times L_{22}\right).\det\left(U_{11} \times U_{22}\right) \\
= \det\left(L_{22}\right).\det\left(U_{11}\right).\det\left(U_{22}\right) \\
= \det\left(L_{22} \times U_{11}\right).\det\left(U_{22}\right)
\end{gather*}

If $A^{-1}$ exists $(\det\left(A\right) \neq 0)$: 

$$
\det\left(M\right) = \det\left(D - CA^{-1}B\right).\det\left(A\right)
$$

Back to the main problem, considering that $Z \in \mathbb{R}^k$ and $Y \in \mathbb{R}^q$ we deifne the function $f_{Y|Z}$ as follows:

$$
f_{Z,Y} = \mathcal{N}( \begin{bmatrix}
   z \\
   y
\end{bmatrix} | m,V) = \frac{1}{\sqrt{(2\pi)^{k+q}\det(V)}} \exp\left( -\frac{1}{2} \left(\begin{bmatrix}
   z \\
   y
\end{bmatrix} - m\right)^\top V^{-1} \left(\begin{bmatrix}
   z \\
   y
\end{bmatrix} - m\right) \right)
$$

Where

$$
V = \begin{bmatrix}
   \Sigma_{Z} & \Sigma_{Z}W^\top \\
   W\Sigma_{Z} & \Sigma_{Y|Z}+W\Sigma_{Z}W^\top
\end{bmatrix} \in \mathbb{R}^{(k+q)\times(k+q)}
\quad , \quad
m = \begin{bmatrix}
   \mu_{Z} \\
   W\mu_{Z}+b
\end{bmatrix} \in \mathbb{R}^{k+q}
$$

Firstly, we try to calculate $\det(V)$ using the lemma that we just proved:

$$
\begin{gather*}
\det(V) = \det(\Sigma_{Y|Z}+W{Z}W^\top - W\Sigma_{Z}\Sigma_{Z}^{-1}\Sigma_{Z}W^\top).\det(\Sigma_{Z}) \\ 
= \det(\Sigma_{Y|Z}+W{Z}W^\top - W\Sigma_{Z}W^\top).\det(\Sigma_{Z}) \\ 
= \det(\Sigma_{Y|Z}).\det(\Sigma_{Z})
\end{gather*}
$$

Secondly, we try to find the answer to $V^{-1}$. One can show that the inverse of the matrix M discussed previously can be written as:

$$
M^{-1} = \begin{bmatrix}
   A^{-1}+A^{-1}BF^{-1}CA^{-1} & -A^{-1}BF^{-1} \\
   -F^{-1}CA^{-1} & F^{-1}
\end{bmatrix} \quad , \quad F = D-CA^{-1}B
$$

Now we apply this to the matrix V:

$$
F = \Sigma_{Y|Z}+W\Sigma_{Z}W^\top-W\Sigma_{Z}\Sigma_{Z}^{-1}\Sigma_{Z}W^\top = \Sigma_{Y|Z}
$$
$$
V^{-1} = \begin{bmatrix}
   \Sigma_{Z}^{-1}+\Sigma_{Z}^{-1}\Sigma_{Z}W^\top \Sigma_{Y|Z}^{-1}W\Sigma_{Z}\Sigma_{Z}^{-1} & -\Sigma_{Z}^{-1}\Sigma_{Z}W^\top \Sigma_{Y|Z}^{-1} \\
   -\Sigma_{Y|Z}^{-1}W\Sigma_{Z}\Sigma_{Z}^{-1} & \Sigma_{Y|Z}^{-1}
\end{bmatrix} = \begin{bmatrix}
   \Sigma_{Z}^{-1}+W^\top \Sigma_{Y|Z}^{-1}W & -W^\top \Sigma_{Y|Z}^{-1} \\
   -\Sigma_{Y|Z}^{-1}W & \Sigma_{Y|Z}^{-1}
\end{bmatrix}
$$


At this point, we are ready to calculate the below term:

$$
\left(\begin{bmatrix}
   z \\
   y
\end{bmatrix} - m\right)^\top V^{-1} \left(\begin{bmatrix}
   z \\
   y
\end{bmatrix} - m\right) = 
\begin{bmatrix}
   z-\mu_Z \\
   y-(W\mu_Z+b)
\end{bmatrix}^\top V^{-1} \begin{bmatrix}
   z-\mu_Z \\
   y-(W\mu_Z+b)
\end{bmatrix}
$$
$$
= \begin{bmatrix}
   z^\top-\mu_Z^\top & y^\top-(\mu_Z^\top W^\top +b^\top)
\end{bmatrix}
\begin{bmatrix}
   \Sigma_{Z}^{-1}+W^\top \Sigma_{Y|Z}^{-1}W & -W^\top \Sigma_{Y|Z}^{-1} \\
   -\Sigma_{Y|Z}^{-1}W & \Sigma_{Y|Z}^{-1}
\end{bmatrix} 
\begin{bmatrix}
   z-\mu_Z \\
   y-(W\mu_Z+b)
\end{bmatrix}
$$
$$
= \begin{bmatrix}
   z^\top\Sigma_{Z}^{-1} + z^\top W^\top \Sigma_{Y|Z}^{-1}W - \mu_Z^\top \Sigma_{Z}^{-1} - y^\top \Sigma_{Y|Z}^{-1}W + b^\top \Sigma_{Y|Z}^{-1}W  &
   -z^\top W^\top \Sigma_{Y|Z}^{-1} + y^\top \Sigma_{Y|Z}^{-1} - b^\top \Sigma_{Y|Z}^{-1}
\end{bmatrix}
\begin{bmatrix}
   z-\mu_Z \\
   y-(W\mu_Z+b)
\end{bmatrix}
$$
$$
= \begin{bmatrix}
   (z-\mu_Z)^\top \Sigma_{Z}^{-1} - \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}W  &
   \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}
\end{bmatrix}
\begin{bmatrix}
   z-\mu_Z \\
   y-(W\mu_Z+b)
\end{bmatrix}
$$
$$
= (z-\mu_Z)^\top \Sigma_{Z}^{-1}(z-\mu_Z) - \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}W(z-\mu_Z) + \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}\left(y - \left(W\mu_Z+b\right)\right)
$$
$$
= (z-\mu_Z)^\top \Sigma_{Z}^{-1}(z-\mu_Z) + \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}\left(y - \left(Wz+b\right)\right)
$$

Using these results, we rewrite the $f_{Z,Y}$:

$$
f_{Z,Y} = \frac{1}{\sqrt{(2\pi)^{k+q}\det(\Sigma_{Y|Z}).\det(\Sigma_{Z})}} \exp\left(-\frac{1}{2}\left( (z-\mu_Z)^\top \Sigma_{Z}^{-1}(z-\mu_Z) + \left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}\left(y - \left(Wz+b\right)\right) \right)\right)
$$
$$
\frac{1}{\sqrt{(2\pi)^{k}\det(\Sigma_{Z})}} \exp\left(-\frac{1}{2}(z-\mu_Z)^\top \Sigma_{Z}^{-1}(z-\mu_Z)\right)
\frac{1}{\sqrt{(2\pi)^{q}\det(\Sigma_{Y|Z})}} \exp\left(-\frac{1}{2}\left(y - \left(Wz+b\right)\right)^\top \Sigma_{Y|Z}^{-1}\left(y - \left(Wz+b\right)\right)\right)
$$
$$
= p_Z(z) \times p_{Y|Z}(y|z) = p_{Z,Y}(z,y)
$$

So $f_{Z,Y}$ is the joint distribution which is a multivariate normal distribution with mean vector m and covariance matrix V:

$$
P_{Z,Y}(z,y) = \mathcal{N}(\mu_{Z,Y},\Sigma_{Z,Y})
$$
$$
\mu_{Z,Y} = \begin{bmatrix}
   \mu_{Z} \\
   W\mu_{Z}+b
\end{bmatrix}
$$
$$
\Sigma_{Z,Y} = \begin{bmatrix}
   \Sigma_{Z} & \Sigma_{Z}W^\top \\
   W\Sigma_{Z} & \Sigma_{Y|Z}+W\Sigma_{Z}W^\top
\end{bmatrix}
$$

According to the theory homework 1, question 8, we can write the conditional distribution of normal random variables having their joint distribution as follows:
$$
P_{Z|Y}(z|y) = \mathcal{N}(\mu_{Z|Y},\Sigma_{Z|Y})
$$
$$
\mu_{Z|Y} = \Sigma_{Z|Y}\left[ W^\top \Sigma_{Y}^{-1}(y-b)+\Sigma_{Z}^{-1}\mu_Z \right]
$$
$$
\Sigma_{Z|Y}^{-1} = \Sigma_{Z}^{-1} + W^\top \Sigma_{Y|Z}^{-1}W \\
\Sigma_{Z|Y} = \Sigma_{Z} -\Sigma_{Z}W^\top (\Sigma_{Y|Z}+W\Sigma_{Z}W^\top )^{-1} W\Sigma_{Z}
$$

# Theory Question 2
Assume that prior distribution is GMM and $p_{Y|Z}$ is a normal distribtion. Is posterior distribution normal? Is it also a GMM? If it is GMM, what are parameters of this GMM if we know parameters of the aforementioned distributions?

Answer:

Assume that the prior has a GMM distribution as below:
$$
p_Z(z) = \sum_k \pi_k p_{Z|k}(z|k) = \sum_k \pi_k \mathcal{N}(\mu_k,\Sigma_k) \quad , \quad \sum_k \pi_k = 1
$$

In the previous question we showed that for each normal component we can easily compute the posterior distribution:

$$
p_{Z|Y,k}(z|y,k) = \frac{p_{Y|Z}(y|z) \times p_{Z|k}(z|k)}{p_{Y|k}(y|k)} = \mathcal{N}(z|\mu_{Z|Y,k},\Sigma_{Z|Y,k})
$$

Where

$$
p_{Y|k}(y|k) = \mathcal{N}(y|\mu_{Y|k},\Sigma_{Y|k}) \\
\mu_{Y|k} = W\mu_{k}+b \\
\Sigma_{Y|k} = \Sigma_{Y|Z}+W\Sigma_{k}W^\top
$$

So

$$
p_{Y|Z}(y|z) \times p_{Z|k}(z|k) = p_{Y|k}(y|k) \times p_{Z|Y,k}(z|y,k) = \mathcal{N}(y|\mu_{Y|k},\Sigma_{Y|k}) \times \mathcal{N}(z|\mu_{Z|Y,k},\Sigma_{Z|Y,k})
$$

Now we try to compute the posterior distribution for the GMM prior:

$$
p_{Z|Y}(z|y) = \frac{p_{Y|Z}(y|z) \times p_{Z}(z)}{p_Y(y)} = \frac{p_{Y|Z}(y|z) \times\sum_k \pi_k p_{Z|k}(z|k)}{\int p_{Y|Z}(y|z) \times\sum_{k'} \pi_{k'} p_{Z|k'}(z|k') dz}
$$
$$
= \sum_k \pi_k \frac{p_{Y|Z}(y|z) \times p_{Z|k}(z|k)}{\sum_{k'} \pi_{k'} \int p_{Y|Z}(y|z) \times p_{Z|k'}(z|k') dz} = \sum_k \frac{\pi_k p_{Y|k}(y|k)}{\sum_{k'} \pi_{k'} p_{Y|k'}(y|k')} \times p_{Z|Y,k}(z|y,k)
$$
$$
= \sum_k (\pi_{new})_k \times \mathcal{N}(z|\mu_{Z|Y,k},\Sigma_{Z|Y,k})
$$

So the posterior’s distribution is a GMM itself with parameters below:

$$
\mu_{Z|Y,k} = \Sigma_{Z|Y,k}\left[ W^\top \Sigma_{Y}^{-1}(y-b)+\Sigma_{k}^{-1}\mu_{k} \right]
$$
$$
\Sigma_{Z|Y,k}^{-1} = \Sigma_{k}^{-1} + W^\top \Sigma_{Y|Z}^{-1}W \\
\Sigma_{Z|Y,k} = \Sigma_{k} -\Sigma_{k}W^\top (\Sigma_{Y|Z}+W\Sigma_{k}W^\top )^{-1} W\Sigma_{k}
$$
$$
(\pi_{new})_k = \frac{\pi_k p_{Y|k}(y|k)}{\sum_{k'} \pi_{k'} p_{Y|k'}(y|k')} = \frac{\pi_k \mathcal{N}(y|\mu_{Y|k},\Sigma_{Y|k})}{\sum_{k'} \pi_{k'} \mathcal{N}(y|\mu_{Y|k},\Sigma_{Y|k})}
$$

And also we can show that $(\pi_{new})_k$ coefficients are valid as thier sum over k is equal to 1:

$$
\sum_k (\pi_{new})_k = \sum_k \frac{\pi_k p_{Y|k}(y|k)}{\sum_{k'} \pi_{k'} p_{Y|k'}(y|k')} = \frac{\sum_k \pi_k p_{Y|k}(y|k)}{\sum_{k'} \pi_{k'} p_{Y|k'}(y|k')} = 1
$$


# Theory Question 3
Answer:

A single gaussian distribution is simple but can fail easily due to its low capacity because it assumes only one mean and variance for all samples. 

On the other hand, GMM can be a good choice because of the nature of the problem. In this problem we’re working with handwritings. A good assumption about handwritings is that we can decompose each image to smaller patches and we can divide these patches into groups like lines or curves in different angles and etc. with each group having similar core shapes with some variations.

This is exactly what GMM tries to do by assuming the model to be a mixture of different gaussian distributions with different coefficients. So, we prefer GMM for the distribution of Z.
